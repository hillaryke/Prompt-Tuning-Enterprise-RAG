```python
from langchain_openai import OpenAIEmbeddings
from langchain.evaluation import load_evaluator
```

##### Turn apple into a vector embedding

```python
embedding_function = OpenAIEmbeddings()
vector = embedding_function.embed_query("apple")
```

```python
print(vector)
print(len(vector))
```

##### Compare embedding distance and similarity

```python
# Run an evaluation to compare the distance between apple and orange
evaluator = load_evaluator("pairwise_embedding_distance")
apple_vs_orange = evaluator.evaluate_string_pairs(prediction="apple", prediction_b = "orange")
apple_vs_beach = evaluator.evaluate_string_pairs(prediction="apple", prediction_b = "beach")
apple_vs_apple = evaluator.evaluate_string_pairs(prediction="apple", prediction_b = "apple")
apple_vs_google = evaluator.evaluate_string_pairs(prediction="apple", prediction_b = "google")
apple_vs_iphone = evaluator.evaluate_string_pairs(prediction="apple", prediction_b = "iphone")
apple_vs_ios = evaluator.evaluate_string_pairs(prediction="apple", prediction_b = "ios")
apple_vs_bus = evaluator.evaluate_string_pairs(prediction="apple", prediction_b = "bus")
print(apple_vs_orange)
print(apple_vs_beach)
print(apple_vs_apple)
print(apple_vs_google)
print(apple_vs_iphone)
print(apple_vs_ios)
print(apple_vs_bus)
```
